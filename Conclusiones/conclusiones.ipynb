# Análisis de Sensibilidad y Conclusiones Estratégicas

Este análisis se basa en la implementación del modelo de planificación de rutas nacionales con estrategias de recarga para LogistiCo. A continuación, se presentan los escenarios de sensibilidad analizados y las conclusiones que pueden orientar decisiones estratégicas para la compañía.

---

##  Análisis de Sensibilidad

### Escenario 1: Aumento de precios en estaciones clave (+20 %)

*Descripción:*  
Se incrementan en un 20 % los precios de combustible en estaciones estratégicas de alto flujo

*Impacto observado:*
- Las rutas óptimas cambian hacia estaciones con precios más bajos, incluso si estas implican desviaciones y un mayor recorrido
por parte del vehículo al reducir costos.
- El costo total del transporte aumenta dependiendo de la estación alternativa que se haya escogido. El aumento del costo total suele ser menor al incremento de los precios de combustible.
- Se reduce el uso de las estaciones a las que se les aumentaron los precios de su combustible.

---

### Escenario 2: Reducción de autonomía por deterioro de camiones

*Descripción:*  
La autonomía de los camiones se reduce en un 25 %, simulando desgaste mecánico.

*Impacto observado:*
- Aumenta el número de paradas de recarga por ruta.
- Se incrementan los costos operativos debido a las paradas adicionales para recargar combustible.
- La necesidad de estaciones intermedias se vuelve importante en rutas de baja cobertura. Puede que la autonomía de los camiones no sea suficiente para cubrir alguna ruta.

---

### Escenario 3: Inclusión o exclusión de estaciones estratégicas

*Descripción:*  
Se evalúa el impacto de cerrar temporalmente o agregar estaciones de recarga en puntos clave.

*Impacto observado:*
- *Exclusión:* Puede incrementar el costo total, además de obligar a tomar rutas menos eficientes con recorridos más largos.
- *Inclusión:* Puede reducir el costo total al facilitar trayectos más directos sin tantas paradas de recarga de combustible.

---

##  Conclusiones Estratégicas

### ¿Dónde debería LogistiCo establecer acuerdos con estaciones para minimizar costos?

*Conclusión:*  
LogistiCo debería priorizar acuerdos comerciales con estaciones ubicadas en:

- Zonas intermedias entre grandes centros urbanos (ej. entre Medellín y Bogotá).
- Intersecciones de rutas.
- Estaciones con precios bajos o descuentos por volumen.


### ¿Qué tipo de camiones son más eficientes según el patrón de demanda?

*Conclusión:*  
Camiones de *autonomía extendida* y *peso optimizado* son más eficientes en rutas largas o con restricciones municipales. Los camiones con menor autonomía requieren más paradas y sufren más por estaciones con precios altos o muy lejanas del camino óptimo entre ciudades.

---

### ¿Cómo afectan los peajes variables la asignación óptima de rutas?

*Conclusión:*  
Peajes elevados pueden hacer preferibles rutas más largas si implican menor costo total. Además, si estos peajes tienen tarifas elevadas por peso del camión, elevarán mucho más los costos. Son preferibles los peajes con precios estables para reducir costos y que la ruta óptima sea la más directa entre el origen y el destino.
